Создание DataFrame

```python
import pandas as pd
```

```python
# Cпособ 1
countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'area': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})
display(countries_df)
```

```python
countries_df.index = ['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ']
display(countries_df)
```

```python
#Способ 2
countries_df = pd.DataFrame(
    data = [
        ['Англия', 56.29, 133396],
        ['Канада', 38.05, 9984670],
        ['США', 322.28, 9826630],
        ['Россия', 146.24, 17125191],
        ['Украина', 45.5, 603628],
        ['Беларусь', 9.5, 207600],
        ['Казахстан', 17.04, 2724902]
    ],
    columns= ['country', 'population', 'area'],
    index = ['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ']
)
display(countries_df)
```

Параметр Axis = 0 - движение по строкам (в рамках одного столбца)
Параметр Axis = 1 - движение по столбцам (в рамках одной строки)

```python
# Среднее по строкам:
countries_df.mean(axis=0, numeric_only=True) #numeric_only в значение True = вести расчёт только по числовым столбцам
```

```python
# Среднее по столбцам:
countries_df.mean(axis=1, numeric_only=True)
```

```python
income = [478, 512, 196]
expenses = [156, 130, 270]
years = [2018, 2019, 2020]

def create_companyDF(income, expenses, years):
    df = pd.DataFrame({
    'income': income,
    'expenses': expenses},
     index = years)
    return df
df = create_companyDF(income, expenses, years)
display(df)
```

```python
df.index.tolist() #выдает индексы в виде списка
df.index.tolist()
```

```python
def get_profit(df, year):
    if year in df.index.tolist():
        profit = df.loc[year, 'income']-df.loc[year, 'expenses']
        return profit
    else:
        return None
get_profit(df, 2020)    
    
```
