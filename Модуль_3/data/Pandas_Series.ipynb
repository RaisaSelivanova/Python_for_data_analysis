Создание series

```python
countries = pd.Series(
    data = ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    index = ['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ'],
    name = 'countries'
)
display(countries)
```

```python
countries = pd.Series({
    'UK': 'Англия',
    'CA': 'Канада',
    'US' : 'США',
    'RU': 'Россия',
    'UA': 'Украина',
    'BY': 'Беларусь',
    'KZ': 'Казахстан'},
    name = 'countries'
)
display(countries)
```

```python
names = ['chlorhexidine', 'cyntomycin', 'afobazol'] 
counts = [15, 18, 7]
def create_medications(names, counts):
    medications = pd.Series(data = counts, index = names, name = 'medications')
    return medications
medications = create_medications(names, counts)
display(medications)
    
```

```python
def get_percent(medications, name):
    count_name = medications.loc[name]
    part_name = count_name*100/sum(medications.iloc[:])
    return part_name
get_percent(medications, 'chlorhexidine')
```
