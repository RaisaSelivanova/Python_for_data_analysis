```python
import pandas as pd
ratings_movies = pd.read_csv('C:/Users/U_M1P8G/Desktop/IDE/Python_for_data_analysis/Модуль_5/data/ratings_movies.csv', sep=',')
```

Для решения задач нам понадобится выделить из признака title год выпуска фильма. Для этого напишем функцию get_year_release(arg).

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg)
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
display(ratings_movies)
```

```python
ratings_movies['year_release'].isna().sum()
```

```python
# Фильтруем фильмы 1999 года
ratings_movies_filtered = ratings_movies[ratings_movies['year_release'] == 1999]

# Группируем по названию и находим средний рейтинг
movie_ratings_1999 = (ratings_movies_filtered
                      .groupby('title')['rating']
                      .mean()
                      .sort_values(ascending=True))

# Выводим результат
print("Фильмы 1999 года с наименьшими средними оценками:")
print(movie_ratings_1999.head(10))

# Получаем фильм с наименьшей оценкой
lowest_rated_movie = movie_ratings_1999.index[0]
lowest_rating = movie_ratings_1999.iloc[0]

print(f"\nФильм с наименьшей средней оценкой: {lowest_rated_movie}")
print(f"Средняя оценка: {lowest_rating:.2f}")
```
